```python
# Import Dependencies
from bs4 import BeautifulSoup as soup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

```

```python
# Create Webdriver
service = Service('/Users/cameronlee/Desktop/DataClass/Spotify-Popularity-Predictor/chromedriver-mac-x64/chromedriver')
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

```

```python
# Visit Website
url = "https://kworb.net/spotify/artists.html"
driver.get(url)
```

```python
# Create Beautiful Soup Object
html = driver.page_source
artist_soup = soup(html, 'html.parser')

print(artist_soup)
```

```python
# Scrape the Column Titles
table = artist_soup.find('table')
header = table.find_all('th')
column = []

for head in header:
    column.append(head.text.strip())

print(column)
```

```python
# Scrape Table Data
artist_list = []
html = driver.page_source
artist_soup = soup(html, 'html.parser')
table = artist_soup.find('table')
artist_data = table.find_all('tr')
for row in artist_data:
    row_data = []
    for col in row.find_all('td'):
        row_data.append(col.text.strip())
    artist_list.append(row_data)

artist_list
```

```python
# Close Browser
driver.quit()
```

```python
# Save as DataFrame
artist_df = pd.DataFrame(artist_list, columns = column)
```

```python
# Display DataFrame
artist_df
```

```python
# Clean DataFrame
artist_df = artist_df.drop(0)
artist_df = artist_df.reset_index(drop = True)
artist_df
```

```python
# Save DataFrame to CSV
artist_df.to_csv('top_artist.csv')
```
